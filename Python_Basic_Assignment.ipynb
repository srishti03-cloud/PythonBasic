2) What is an interpreter in Python?

An interpreter is a program that executes Python code line by line instead of converting the whole program into machine code at once.

3) What are pre-defined keywords in Python?

Pre-defined keywords are reserved words that have special meanings in Python and cannot be changed.

Examples: if,else,elif,for,while,etc

4) Can keywords be used as variable names?

No,Keywords cannot be used as variable names because they are reserved by Python.

5) What is mutability in Python?

Mutability means whether an object’s value can be changed after it is created.
Mutable → Can be changed (List, Dictionary, Set)

6)Why are lists mutable, but tuples are immutable?

Lists are mutable because they are designed for situations where data may change.

Tuples are immutable to ensure data safety and faster performance.

7)What is the difference between == and is operators in Python?

== → Compares values

is → Compares memory location (identity)


```python
a = [1, 2]
b = [1, 2]

a == b   # True (values are same)
a is b   # False (different memory location)
```

Logical operators are used to combine conditional statements.

or → True if at least one condition is True

not → Reverses the condition

11)What is the purpose of conditional statements in Python?

Conditional statements allow a program to make decisions based on conditions.

12)How does the elif statement work?

elif means “else if”.

It checks another condition if the previous if condition is False.

```python
marks = 75

if marks >= 90:
    print("A Grade")
elif marks >= 60:
    print("B Grade")
else:
    print("Fail")
```

14) Describe a scenario where a while loop is more suitable than a for loop.

A while loop is better when the number of iterations depends on a condition that may change dynamically.

```python
password = ""
while password != "admin123":
    password = input("Enter password: ")
```

```python
#Function to demonstrate mutable and immutable arguments
def modify(num, lst):
    num = num + 5
    lst.append(4)
    print("Inside function:", num, lst)

number = 10
my_list = [1, 2, 3]

modify(number, my_list)

print("Outside function:", number, my_list)
```

```python
#Program to find sum of even numbers between 1 and 50
sum = 0

for i in range(1, 51):
    if i % 2 == 0:
        sum += i

print("Sum of even numbers:", sum)
```

```python
#Reverse a string using while loop
text = input("Enter a string: ")

i = len(text) - 1

while i >= 0:
    print(text[i], end="")
    i -= 1
```

```python
#Factorial using while loop
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact = fact * i
    i += 1

print("Factorial is:", fact)
```
